The url to get my starred repository.

```python
last_resp = ''
first_url_to_get = 'https://api.github.com/user/starred'
first_url_resp = requests.get(first_url_to_get, auth = (github_uname,github_token))
```

The below code will retrieve all the repositories that are starred by me.

```python
name_list = []
url_list = []
owner_list = []
for repo in json.loads(first_url_resp.text):
  name_list.append(repo['name'])
  owner_list.append(repo['owner']['login'])
  url_list.append(repo['owner']['html_url'])
last_resp = first_url_resp
while last_resp.links.get('next'):
    next_url_to_get = last_resp.links['next']['url']
    next_url_resp = requests.get(next_url_to_get, auth= (github_uname, github_token))
    for repo in json.loads(next_url_resp.text):
      name_list.append(repo['name'])
      owner_list.append(repo['owner']['login'])
      url_list.append(repo['owner']['html_url'])
    last_resp = next_url_resp
```

  Now we will get all the repositories these guys have starred in their profile, below code will retrieve all the repositories thatt are starred by owner of the repositories that I have starred in my profile.

```python
name_list_star = []
url_list_star = []
owner_list_star = []
starred_by_list_star = []
user_dict = {}
for owner in owner_list:
  temp_name_list_star = []
  first_url_to_get = 'https://api.github.com/users/' + owner + '/starred'
  first_url_resp = requests.get(first_url_to_get, auth = (github_uname, github_token))
  last_resp = first_url_resp
  all_starred_rep = json.loads(first_url_resp.text)

  for repo in json.loads(first_url_resp.text):
    #if(repo['name'] not in name_list_star): #to avoid same repository that would be bookmarked by diffrent users.
    name_list_star.append(repo['name'])
    owner_list_star.append(repo['owner']['login'])
    url_list_star.append(repo['owner']['html_url'])
    starred_by_list_star.append(owner)
    temp_name_list_star.append(repo['name'])
  while last_resp.links.get('next'):
    next_url_to_get = last_resp.links['next']['url']
    next_url_resp = requests.get(next_url_to_get, auth= (github_uname, github_token))
    last_resp = next_url_resp
    for repo in json.loads(next_url_resp.text):
      name_list_star.append(repo['name'])
      owner_list_star.append(repo['owner']['login'])
      url_list_star.append(repo['owner']['html_url'])
      starred_by_list_star.append(owner)
      temp_name_list_star.append(repo['name'])
  user_dict[owner] = temp_name_list_star
```

```python
from scipy.stats import pearsonr
similarity_score = {}
for i in range(len(df)):
 ss = pearsonr(df.iloc[-1,:], df.iloc[i,:])
 similarity_score.update({i: ss[0]})
sf = pd.Series(similarity_score).to_frame('similarity')
```

```python
sf
```

```python
sf = sf.sort_values('similarity', ascending=False) #this will sort and keep the index accordingly..
```

```python
sf
```

```python
#ignoring the first one with score of 1, as it is our own only..
recommendation = sf.index[1:4] #getting top 3 recommendations..
```

```python
#ignoring the first one with score of 1, as it is our own only..
print("Top 3 recommendations :\n1.", df.index[recommendation[0]], "\n2.", df.index[recommendation[1]], "\n3.", df.index[recommendation[2]])
```
